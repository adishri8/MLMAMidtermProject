To quantify activity levels from Empatica (E4) accelerometer data, convert raw $x, y, z$ values to gravity units ($1/64g$), calculate the Vector Magnitude (VM) to account for3D movement ($VM = \sqrt{x^2+y^2+z^2}$), and calculate the standard deviation or mean absolute difference over time-based epochs (e.g., 1-second or 1-minute intervals). [1, 2, 3]  
Steps to Quantify Empatica E4 Activity Data 

• 1. Data Preprocessing (ACC.csv): 

	• Convert Units: The E4 records acceleration in $\pm2g$ range, where data units are $1/64g$. Divide the raw $x, y, z$ values by 64 to convert them into $g$-units. 
	• Data Structure: The  file provides three columns representing the $x, y, \text{and} z$ axes, usually sampled at 32 Hz or 4 Hz. 

• 2. Calculate Resultant Acceleration (Vector Magnitude): 

	• To get a single metric of movement intensity, compute the magnitude of the 3D acceleration vectors to negate device orientation. 
	• Formula: $VM = \sqrt{x^2 + y^2 + z^2}$. 
	• Note: Some approaches subtract 1g from the result to remove the gravity component when the device is at rest. 

• 3. Feature Extraction (Quantification): 

	• Standard Deviation (SD): Calculate the standard deviation (STD) for each axis or the VM over a specific timeframe (epoch), such as 1 second. A higher SD indicates higher intensity. 
	• Mean Absolute Difference (MAD): Calculate the mean absolute difference from the mean, or use simple averages of the VM over 1-minute epochs to calculate activity counts. 

• 4. Activity Intensity Classification: 

	• Sedentary/Low Activity: Low Standard Deviation (close to zero). 
	• High Activity: High Standard Deviation. 
	• Empatica Mate App: The Empatica Support website (https://support.empatica.com/hc/en-us/articles/115000184666-Mate-analysis-of-physical-activity) app automatically classifies movement into walking, running, and sports, with higher intensity bars indicating greater movement intensity. [1, 2, 3, 4, 5, 6]  

Common Pitfalls: 

• Sensor Noise: Raw accelerometer data can be noisy; applying a low-pass filter can help isolate body movement from high-frequency vibration. 
• Calibration: The Empatica E4 automatically calibrates during the first 15 seconds of a session. [3, 7, 8]
