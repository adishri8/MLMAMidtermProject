```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('merged_data.csv.zip', compression='zip', dtype={'id': str})

for id, df_nurse in df.groupby('id'):
    df_nurse['time'] = (np.arange(len(df_nurse)) * 0.03).round(2)
    df_nurse['acc_mag'] = np.sqrt(df_nurse['X']**2 + df_nurse['Y']**2 + df_nurse['Z']**2)
    # df['label'] = df['label'].replace({1.0: 2.0})
    df_nurse.drop(columns=['datetime', 'X', 'Y', 'Z'], inplace=True)
    df_nurse = df_nurse[['time', 'acc_mag', 'EDA', 'HR', 'TEMP', 'label']]
    df_nurse.to_csv(f'data/Eric/processed_nurse_{id}.csv', index=False)
```
