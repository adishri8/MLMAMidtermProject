# Compiled Results Table

This notebook loads and displays the compiled results table saved in `outputs/compiled_results_table.csv`.

```python
import os
import pandas as pd
from IPython.display import display, Markdown

rel_path = 'outputs/compiled_results_table.csv'
abs_path = '/Users/adityashrinivasan/Documents/Semester 6/Machine Learning for Medical Applications/MLMAMidtermProject/outputs/compiled_results_table.csv'

print('Notebook cwd:', os.getcwd())
print('Outputs dir listing:')
try:
    print(os.listdir('outputs'))
except Exception as e:
    print('Could not list outputs directory:', e)

# Try relative read first, fall back to absolute path
try:
    df = pd.read_csv(rel_path)
    display(Markdown(f'**Loaded (relative):** `{rel_path}`'))
except FileNotFoundError:
    try:
        df = pd.read_csv(abs_path)
        display(Markdown(f'**Loaded (absolute):** `{abs_path}`'))
    except Exception as e:
        raise FileNotFoundError(f'Neither {rel_path} nor {abs_path} could be read: {e}')

# Keep only the first five columns
df = df.iloc[:, :5].copy()
display(df)
```
